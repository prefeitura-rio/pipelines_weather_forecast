```python
import datetime
from time import sleep

def timeout_flow(timeout_minutes: int =30):    
    """
    Stop flow if it exceeds timeout_minutes
    """
    start_time = datetime.datetime.now(datetime.timezone.utc)
    while True:
        elapsed_time = datetime.datetime.now(datetime.timezone.utc) - start_time
        if elapsed_time > datetime.timedelta(seconds=timeout_minutes):
            print("Tempo limite excedido. Encerrando o fluxo.")
            break
            # raise Failed(f"Stop flow after {timeout_minutes} minutes")
        sleep(10)

timeout_flow()
```
